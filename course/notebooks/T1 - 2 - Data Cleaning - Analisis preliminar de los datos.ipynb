Los valores que faltan en un data set pueden venir por dos razones:
* Extracción de los datos (cuando se extraen datos de una base de datos a nuestro datos, pueden haber incompatibilidades y quedan como nulos, porque no son capaces de incorparar al fichero, se puede corregir este proceso mejorando el proceso de extraccion o haciendo una opteccion de datos mucha mas precisas)
* Recolección de los datos (en algunos casos no se guardan en la base de datos, ejemplo una encuesta, proque no quieran contestar algunas pregunctas, es decir que ni siquiera se pueda extraer de la base de datos)



### Borrados de valors que faltan


```python
# se pueden borrar filas o columnas
```

```python
#borrado de fila
# axis= es el eje horizontal --> fila, axis = 1 elimina la columna
#how que filas tiene que borrar. all --> sera borrada si todas sus columnas contiene sus valores nulos
data.dropna(axis=0, how ='all')
```

```python
# how = any ---> eliminar si alguna de las columnas posea un valor nulo
data2 = data
data2.dropna(axis=0, how ='any')
```

### Inferir los datos - COmputo de los valores faltantes

```python
# consiste en remplazar valores por otro, un promedio, rellenar el valor con cero, ... LA eleccion del mejor metodo depende del contexto
```

```python
# rellanar los valores faltantes con cero
data3 = data
data3.fillna(0)
```

```python
# se puede remplazar los valores faltantes por una palabra
data4 = data
data4.fillna('Desconocido')
```

```python
# remplaxar dependiendo de la columna
data5 = data
data5['body'] = data5['body'].fillna(0)
data5['home.dest'] = data5['home.dest'].fillna('Desconocido')
data.head()
```

```python
# se puede reemplazar por la media(el promedio, excluyendo los valores q faltan)
pd.isnull(data5['age']).values.ravel().sum()
```

```python
# premedio . mean()
data5['age'].fillna(data5['age'].mean())
```

```python
# se lo puede reeemplazar con el valor mas cerca conocido que encuentre antes del mismo y despues del mismo
# metodo ffill ----> forwardFill: va hacia adelante
data5['age'].fillna(method = 'ffill')
```

```python
# metodo backfill ----> backFill: va hacia atras
data5['age'].fillna(method = 'backfill')
```

#### funcion para create dummy

```python

def createDummies(df, var_name):
    dummy = pd.get_dummies(df[var_name], prefix=var_name)
    df = df.drop(var_name, axis = 1)
    df = pd.concat([df, dummy ], axis = 1)
    return df
```
